```python
# prompt: get all folder have format "/content/drive/MyDrive/filtered_clips/{meeting_id}-{person_id}/clip_{start_time}_{end_time}.mp4" and create a pandas dataframe

import os
import pandas as pd

folder_path = '/content/drive/MyDrive/filtered_clips'

data = []
for root, dirs, files in os.walk(folder_path):
  for dir_name in dirs:
    # Check if the directory name matches the format {meeting_id}-{person_id}
    if '-' in dir_name:
      meeting_id, person_id = dir_name.split('-', 1)
      # Check if the directory contains mp4 files in the specified format
      dir_full_path = os.path.join(root, dir_name)
      for file_name in os.listdir(dir_full_path):
        if file_name.endswith('.mp4') and file_name.startswith('clip_'):
          parts = file_name.replace('.mp4', '').split('_')
          if len(parts) == 3:
            try:
              start_time = int(parts[1])
              end_time = int(parts[2])
              full_file_path = os.path.join(dir_full_path, file_name)
              data.append({
                  'folder_path': dir_full_path,
                  'meeting_id': meeting_id,
                  'person_id': person_id,
                  'file_name': file_name,
                  'start_time': start_time,
                  'end_time': end_time,
                  'full_file_path': full_file_path
              })
            except ValueError:
              # Handle cases where start_time or end_time are not integers
              pass

df = pd.DataFrame(data)
```

```python
# prompt: meeting_to_person have each items contain the list of people participate in the meetting, for each row in the df, create reference row that randomly select 3 other member that join the meeting

import random

def get_reference_persons(row):
  meeting_id = row['meeting_id']
  person_id = row['person_id']
  participants = meeting_to_person.get(meeting_id, [])

  # Exclude the current person from the list of potential references
  potential_references = [p for p in participants if p != person_id]

  # Randomly select up to 3 distinct reference persons
  num_references = min(3, len(potential_references))
  if num_references > 0:
    return random.sample(potential_references, num_references)
  else:
    return []

df['reference_persons'] = df.apply(get_reference_persons, axis=1)

print(df[['meeting_id', 'person_id', 'reference_persons']])
```

```python
import cv2
import os
def save_segments_to_clips(video_path, segments, output_dir):
    os.makedirs(output_dir, exist_ok=True)

    cap = cv2.VideoCapture(video_path)
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    width, height = int(cap.get(3)), int(cap.get(4))
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    clip_paths = []

    for i, (start, end) in enumerate(segments):
        cap.set(cv2.CAP_PROP_POS_MSEC, start * 1000)
        out_path = os.path.join(output_dir, f"clip_{start}_{end}.mp4")
        out = cv2.VideoWriter(out_path, fourcc, fps, (width, height))
        while cap.get(cv2.CAP_PROP_POS_MSEC) < end * 1000:
            ret, frame = cap.read()
            if not ret:
                break
            out.write(frame)
        out.release()
        clip_paths.append(out_path)

    cap.release()
    return clip_paths
```
