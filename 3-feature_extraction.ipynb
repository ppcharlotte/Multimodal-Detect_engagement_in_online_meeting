```python
# extract segment time
def extract_segment_time_from_filename(filename):
    match = re.search(r'(clip|segment)_(\d+)_(\d+)\.mp4', filename)
    if match:
        return int(match.group(2)), int(match.group(3))
    return None, None

# get fps
def get_fps(video_path):
    cap = cv2.VideoCapture(video_path)
    fps = cap.get(cv2.CAP_PROP_FPS)
    cap.release()
    return fps

# read keypoint
def read_keypoint_folder(folder_path):
    keypoint_data = []
    files = sorted([f for f in os.listdir(folder_path) if f.endswith('.json')])
    for fname in files:
        fpath = os.path.join(folder_path, fname)
        try:
            with open(fpath, 'r') as f:
                keypoint_data.append(json.load(f))
        except:
            keypoint_data.append({})
    return keypoint_data
```

```python
def estimate_head_orientation(keypoints):
    """
    Estimate head orientation (yaw, pitch, roll) from 2D facial keypoints.

    Parameters:
    - keypoints: OpenPose keypoints dictionary

    Returns:
    - orientation_vector: [yaw, pitch, roll] in degrees
    """
    try:
        pose_keypoints = keypoints['people'][0]['pose_keypoints_2d']
        keypoints_array = np.array(pose_keypoints).reshape(-1, 3)

        # Extract facial keypoints (indices from OpenPose face model)
        nose = keypoints_array[0][:2]
        left_eye = keypoints_array[16][:2]
        right_eye = keypoints_array[15][:2]

        # Check confidence scores
        confidences = [keypoints_array[i][2] for i in [0, 15, 16]]
        if any(conf <= 0.3 for conf in confidences):  # Increased threshold
            return None

        # Calculate eye center and eye line
        eye_center = (left_eye + right_eye) / 2.0
        eye_line = right_eye - left_eye
        eye_distance = np.linalg.norm(eye_line)

        # Avoid division by zero
        if eye_distance < 1e-6:
            print("Warning: Eyes too close together")
            return None

        # ROLL: Rotation around Z-axis (head tilt)
        # Angle of the line connecting the eyes
        roll = math.degrees(math.atan2(eye_line[1], eye_line[0]))

        # YAW: Rotation around Y-axis (left-right head turn)
        # Based on horizontal offset of nose from eye center
        # Normalize by eye distance to make it scale-invariant
        nose_eye_offset_x = (nose[0] - eye_center[0]) / eye_distance
        # Clamp to reasonable range and scale
        nose_eye_offset_x = np.clip(nose_eye_offset_x, -0.5, 0.5)
        yaw = math.degrees(math.asin(nose_eye_offset_x * 2))  # Scale up the sensitivity

        # PITCH: Rotation around X-axis (up-down head tilt)
        # Based on vertical offset of nose from eye center
        # Normalize by eye distance
        nose_eye_offset_y = (nose[1] - eye_center[1]) / eye_distance
        # Clamp to reasonable range and scale
        nose_eye_offset_y = np.clip(nose_eye_offset_y, -1.0, 1.0)
        pitch = math.degrees(math.atan(nose_eye_offset_y))

        return [yaw, pitch, roll]

    except Exception as e:
        print(f"Error in estimate_head_orientation: {e}")
        return None
```

```python
def get_average_head_orientation(segment, fps, keypoint_list, log=False):
    """
    Calculate average head orientation over a time segment from a list of keypoints.

    Parameters:
    - segment: [start_time, end_time] in seconds
    - fps: frames per second of the video
    - keypoint_list: list of keypoint dictionaries for each frame

    Returns:
    - average_orientation: [avg_yaw, avg_pitch, avg_roll] in degrees
    - valid_frames: number of frames with valid orientation data
    - orientation_std: [std_yaw, std_pitch, std_roll] standard deviation
    """
    start_time, end_time = segment

    # Convert time segment to frame indices
    start_frame = int(start_time * fps)
    end_frame = int(end_time * fps)

    # Ensure frame indices are within bounds
    start_frame = max(0, start_frame)
    end_frame = min(len(keypoint_list) - 1, end_frame)

    if start_frame >= end_frame:
        print(f"Warning: Invalid segment [{start_time}, {end_time}] seconds")
        return None, 0, None

    valid_orientations = []
    failed_frames = 0

    # Process each frame in the segment
    for frame_idx in range(start_frame, end_frame + 1):
        if frame_idx < len(keypoint_list):
            keypoints = keypoint_list[frame_idx]

            # Skip if no people detected in frame
            if not keypoints.get('people') or len(keypoints['people']) == 0:
                failed_frames += 1
                continue

            # Estimate head orientation for this frame
            orientation = estimate_head_orientation(keypoints)

            if orientation is not None:
                valid_orientations.append(orientation)
            else:
                failed_frames += 1

    # Check if we have enough valid data
    valid_frames = len(valid_orientations)

    if valid_frames == 0:
        print(f"Warning: No valid head orientations found in segment [{start_time}, {end_time}]")
        return None, 0, None

    if valid_frames < 3:
        print(f"Warning: Only {valid_frames} valid frames found in segment")

    # Convert to numpy array for easier computation
    orientations_array = np.array(valid_orientations)

    # Handle angle wraparound for roll (can wrap around ±180°)
    # Use circular mean for roll calculation
    roll_radians = np.radians(orientations_array[:, 2])
    roll_sin_mean = np.mean(np.sin(roll_radians))
    roll_cos_mean = np.mean(np.cos(roll_radians))
    avg_roll = np.degrees(np.arctan2(roll_sin_mean, roll_cos_mean))

    # For yaw and pitch, use regular mean (assuming they don't wrap around in typical usage)
    avg_yaw = np.mean(orientations_array[:, 0])
    avg_pitch = np.mean(orientations_array[:, 1])
    average_orientation = [avg_yaw, avg_pitch, avg_roll]

    # Calculate standard deviations
    # For roll, use circular standard deviation
    roll_std = np.degrees(np.sqrt(-2 * np.log(np.sqrt(roll_sin_mean**2 + roll_cos_mean**2))))

    # For yaw and pitch, use regular standard deviation
    yaw_std = np.std(orientations_array[:, 0])
    pitch_std = np.std(orientations_array[:, 1])

    orientation_std = [yaw_std, pitch_std, roll_std]

    # Print summary statistics
    if log:
        print(f"Segment [{start_time:.1f}s, {end_time:.1f}s]: "
            f"Valid frames: {valid_frames}/{end_frame - start_frame + 1}")
        print(f"Average orientation - Yaw: {avg_yaw:.1f}° (±{yaw_std:.1f}°), "
            f"Pitch: {avg_pitch:.1f}° (±{pitch_std:.1f}°), "
            f"Roll: {avg_roll:.1f}° (±{roll_std:.1f}°)")

    return average_orientation, valid_frames, orientation_std
```

```python
#noding detect
from scipy.signal import find_peaks
from scipy.ndimage import gaussian_filter1d

def detect_head_nod(orientations, fps, fast_thresh=3, slow_thresh=6, fast_interval=0.6, slow_interval=1.5, smooth=True):
    """
    使用峰谷配对方式检测点头次数。
    pitch_thresh：最小俯仰角变化幅度（度）
    max_interval：允许点头动作发生的最大时间间隔（秒）
    """
    if len(orientations) < 3:
        return {
            "nod_fast": 0,
            "nod_slow": 0,
            "nod_total": 0,
            "nod_frequency": 0.0,
            "nod_type": "none"
        }

    pitch_series = np.array([o[1] for o in orientations])

    if smooth:
        pitch_series = gaussian_filter1d(pitch_series, sigma=0.5)

    duration = len(pitch_series) / fps

    def detect_nods(thresh, max_interval):
        valleys, _ = find_peaks(-pitch_series, prominence=thresh)
        peaks, _ = find_peaks(pitch_series, prominence=thresh)

        count = 0
        i = j = 0
        while i < len(valleys) and j < len(peaks):
            vi, pj = valleys[i], peaks[j]
            interval = (pj - vi) / fps
            amp = abs(pitch_series[pj] - pitch_series[vi])

            if vi < pj and interval < max_interval and amp >= thresh:
                count += 1
                i += 1
                j += 1
            elif vi >= pj:
                j += 1
            else:
                i += 1
        return count

    fast_nods = detect_nods(fast_thresh, fast_interval)
    slow_nods = detect_nods(slow_thresh, slow_interval)

    total_nods = fast_nods + slow_nods
    frequency = total_nods / duration if duration > 0 else 0

    if total_nods == 0:
        nod_type = 'none'
    elif frequency < 0.5:
        nod_type = 'occasional'
    else:
        nod_type = 'frequent'

    return {
        "nod_fast": fast_nods,
        "nod_slow": slow_nods,
        "nod_total": total_nods,
        "nod_frequency": round(frequency, 3),
        "nod_type": nod_type
    }

#shake head
def detect_head_shake(orientations, fps, yaw_thresh=12, max_interval=1.2, smooth=True):
    """
    使用峰谷配对方式检测摇头次数（基于 yaw 角）。
    yaw_thresh：最小偏航角变化幅度（度）
    max_interval：允许摇头动作发生的最大时间间隔（秒）
    smooth：是否对 yaw 序列进行高斯平滑
    """
    if len(orientations) < 3:
        return 0, 0.0, 'none'

    yaw_series = np.array([o[0] for o in orientations])

    if smooth:
        yaw_series = gaussian_filter1d(yaw_series, sigma=0.6)

    # 找出左转（负峰）和右转（正峰）
    left_peaks, _ = find_peaks(-yaw_series, prominence=yaw_thresh)
    right_peaks, _ = find_peaks(yaw_series, prominence=yaw_thresh)

    shake_events = 0
    i = j = 0
    while i < len(left_peaks) and j < len(right_peaks):
        li, rj = left_peaks[i], right_peaks[j]
        interval = (rj - li) / fps

        if li < rj and interval < max_interval:
            if abs(yaw_series[rj] - yaw_series[li]) >= yaw_thresh:
                shake_events += 1
            i += 1
            j += 1
        elif li >= rj:
            j += 1
        else:
            i += 1

    duration = len(yaw_series) / fps
    shake_freq = shake_events / duration if duration > 0 else 0.0

    if shake_events == 0:
        shake_type = 'none'
    elif shake_freq < 0.5:
        shake_type = 'occasional'
    else:
        shake_type = 'frequent'

    return shake_events, shake_freq, shake_type
```

```python
def analyze_head_movement_patterns(segment, fps, keypoint_list, window_size=1.0):
    """
    Analyze head movement patterns over time within a segment.

    Parameters:
    - segment: [start_time, end_time] in seconds
    - fps: frames per second of the video
    - keypoint_list: list of keypoint dictionaries for each frame
    - window_size: size of sliding window in seconds for movement analysis

    Returns:
    - movement_analysis: dictionary with movement patterns and statistics
    """
    start_time, end_time = segment
    start_frame = int(start_time * fps)
    end_frame = int(end_time * fps)

    # Collect orientations with frame indices
    frame_orientations = {}
    for frame_idx in range(start_frame, end_frame + 1):
        if frame_idx < len(keypoint_list):
            keypoints = keypoint_list[frame_idx]

            if keypoints.get('people') and len(keypoints['people']) > 0:
                orientation = estimate_head_orientation(keypoints)
                if orientation is not None:
                    frame_orientations[frame_idx] = orientation

    if len(frame_orientations) < 2:
        return {}

    # Calculate frame-to-frame differences
    sorted_frames = sorted(frame_orientations.keys())
    angular_velocities = []
    angular_velocities_abs = []
    time_diffs = []

    for i in range(1, len(sorted_frames)):
        prev_frame = sorted_frames[i-1]
        curr_frame = sorted_frames[i]

        prev_orientation = frame_orientations[prev_frame]
        curr_orientation = frame_orientations[curr_frame]

        # Calculate angular differences
        diff = np.array(frame_orientations[curr_frame]) - np.array(frame_orientations[prev_frame])

        # Time difference
        time_diff = (curr_frame - prev_frame) / fps

        # Angular velocities (degrees per second)
        if time_diff > 0:
            angular_velocities.append(diff / time_diff)
            time_diffs.append(time_diff)

    if len(angular_velocities) == 0:
        return {}

    velocities_array = np.array(angular_velocities)

    # Calculate movement statistics
    movement_analysis = {
        'segment': segment,
        'num_valid_transitions': len(angular_velocities),
        'avg_time_between_detections': np.mean(time_diffs),

        # Average angular velocities
        'avg_yaw_velocity': np.mean(np.abs(velocities_array[:, 0])),
        'avg_pitch_velocity': np.mean(np.abs(velocities_array[:, 1])),
        'avg_roll_velocity': np.mean(np.abs(velocities_array[:, 2])),

        # Maximum angular velocities
        'max_yaw_velocity': np.max(np.abs(velocities_array[:, 0])),
        'max_pitch_velocity': np.max(np.abs(velocities_array[:, 1])),
        'max_roll_velocity': np.max(np.abs(velocities_array[:, 2])),

        # Total movement
        'total_yaw_movement': np.sum(np.abs(velocities_array[:, 0]) * np.array(time_diffs)),
        'total_pitch_movement': np.sum(np.abs(velocities_array[:, 1]) * np.array(time_diffs)),
        'total_roll_movement': np.sum(np.abs(velocities_array[:, 2]) * np.array(time_diffs)),

        # Movement variability
        'yaw_velocity_std': np.std(velocities_array[:, 0]),
        'pitch_velocity_std': np.std(velocities_array[:, 1]),
        'roll_velocity_std': np.std(velocities_array[:, 2]),
    }

    # Classify movement intensity
    total_movement = (movement_analysis['total_yaw_movement'] +
                     movement_analysis['total_pitch_movement'] +
                     movement_analysis['total_roll_movement'])

    duration = end_time - start_time
    movement_intensity = total_movement / duration

    if movement_intensity < 10:
        movement_classification = "Static"
    elif movement_intensity < 30:
        movement_classification = "Low movement"
    elif movement_intensity < 60:
        movement_classification = "Moderate movement"
    else:
        movement_classification = "High movement"

    movement_analysis['movement_intensity'] = movement_intensity
    movement_analysis['movement_classification'] = movement_classification

    return movement_analysis
```

```python
def extract_behavior_features(video_path, keypoints_path, segment_id=None):

    if segment_id is None:
        segment_id = f"{os.path.basename(os.path.dirname(video_path))}_{os.path.basename(video_path).replace('.mp4', '')}"

    start, end = extract_segment_time_from_filename(os.path.basename(video_path))
    if start is None or end is None:
        return None

    keypoint_list = read_keypoint_folder(keypoints_path)
    fps = get_fps(video_path)
    start_frame = int(start * fps)
    end_frame = int(end * fps)
    segment = (start, end)

    valid_orientations = []
    hand_raised = []
    hand_near_face = []

    for frame_idx in range(start_frame, end_frame + 1):
        if frame_idx >= len(keypoint_list): break
        keypoints = keypoint_list[frame_idx]
        if not keypoints.get('people') or len(keypoints['people']) == 0:
            continue

        kpts = np.array(keypoints['people'][0]['pose_keypoints_2d']).reshape(-1, 3)

        orientation = estimate_head_orientation(keypoints)
        if orientation:
            valid_orientations.append(orientation)



    nod_result = detect_head_nod(valid_orientations, fps)
    shake_count, shake_frequency, shake_type = detect_head_shake(valid_orientations, fps)


    avg_pose, valid_frames, pose_std = get_average_head_orientation(segment, fps, keypoint_list)
    movement = analyze_head_movement_patterns(segment, fps, keypoint_list)
    print(movement)
    return {
        'segment_id': segment_id,
        'video_path': video_path,
        'start_time': start,
        'end_time': end,
        'duration': end - start,
        'valid_frames': valid_frames,

        # head movement
        'avg_pose': avg_pose,
        'pose_std': pose_std,
        # 'head_movement_intensity': movement['movement_intensity'] if movement else None,
        # 'head_movement_classification': movement['movement_classification'] if movement else None,
        **movement,
        # nodding / shaking
        'nod_count': nod_result["nod_total"],
        'nod_frequency': nod_result["nod_frequency"],
        'nod_type': nod_result["nod_type"],
        'fast_nod_count': nod_result["nod_fast"],
        'slow_nod_count': nod_result["nod_slow"],
        'shake_count': shake_count,
        'shake_frequency': shake_frequency,
        'shake_type': shake_type,
    }
```

```python
# Facial expressions: Track changes in key facial action units associated with attention, understanding, and emotional response
"""
#AU01: inner brow raiser, （thinking,concentrate
#AU02: outer brow raiser, (surprise or ?increased engagement)
#AU04: brow lowerer, (frown, maybe signal focus)
#AU05：upper lid raiser, （widen eyes)
#AU07: lid tightener, (the eyelids are tightened， give the impression of squinting, scrutinizing, or intense focus.)

#### Cognition
#AU10:Upper Lip Raiser, (Slight opening of the mouth during thought)
#AU14: Dimpler,（ambiguous smile
#AU15: Lip Corner Depressor, （confused, especially with AU01+AU04, or AU17)
#AU17: Chin Raiser, (suppression，cognitive effort, nervous)
#AU23: Lip Tightener,(cognitive control, nervousness, emotional restraint)

#### Emotion
#AU06: Cheek raiser (happy, real smile)

#AU12: Lip Corner puller (smile)

#AU25: Lips part (speak or astonished or other possible emotion)

#AU26: Jaw drop 惊讶、惊异(Surprise, astonishment)

#AU45: blink （Fatigue, anxiety, comprehension difficulties, frequent blinking）
"""
```

```python
import os
import subprocess
import pandas as pd
from scipy.signal import find_peaks

#lower threshold for some subtle(low range of motion) AU
AU_THRESHOLD = {
    "AU06_r": 0.35,
    "AU12_r": 0.35,
    "AU14_r": 0.3,
    "AU15_r": 0.35,
    "AU17_r": 0.3,
    "AU23_r": 0.3,
}
DEFAULT_AU_THRESHOLD = 0.6

def extract_au_stats(df, au_col, fps=25):
    threshold = AU_THRESHOLD.get(au_col, DEFAULT_AU_THRESHOLD)
    series = df[au_col]
    above_thresh = series > threshold

    # Frequency: continuous activations frequency

    peaks, _ = find_peaks(series, height=threshold)
    peak_frequency = len(peaks)

    #if frequency == 0:
     #   return None

    # Onset time
    onset_time = None
    if any(above_thresh):
        onset_index = above_thresh.idxmax()
        onset_time = onset_index / fps

    # Duration
    total_duration = above_thresh.sum() / fps

    return {
        f"{au_col}_mean": series.mean(),
        f"{au_col}_std": series.std(),
        f"{au_col}_peak_frequency": peak_frequency,
        f"{au_col}_onset": onset_time if onset_time is not None else -1,
        f"{au_col}_duration": total_duration
    }

def extract_and_analyze_au(video_path, output_dir="/content/openface_output", fps=25):
    """
    extract AU expression and analyse.
    """
    os.makedirs(output_dir, exist_ok=True)

    # 1. extract au features
    command = f'/content/OpenFace/build/bin/FeatureExtraction -f "{video_path}" -aus -2Dfp -pose -out_dir "{output_dir}"'
    print(f"extracting AU : {video_path}")
    subprocess.run(command, shell=True, check=True)

    # 2. read from features file
    csv_files = [f for f in os.listdir(output_dir) if f.endswith('.csv')]
    if not csv_files:
        raise FileNotFoundError("Can't find AU features file.")
    csv_path = os.path.join(output_dir, csv_files[0])

    df = pd.read_csv(csv_path)
    au_r_cols = [col for col in df.columns if col.endswith('_r')]

    if not au_r_cols:
        raise ValueError("No AU_r columns found in the output.")

    # 3. statistical analysis
    result_dict = {}
    for au_col in au_r_cols:
        stats = extract_au_stats(df, au_col, fps=fps)
        result_dict.update(stats)

    return result_dict
```

##### detecting chin rest, touching face, hand raise and support forhead

```python
import pandas as pd
import numpy as np
import cv2
import mediapipe as mp

def extract_pose_wrist(video_path):
    mp_pose = mp.solutions.pose
    pose = mp_pose.Pose(static_image_mode=False)
    cap = cv2.VideoCapture(video_path)

    data = []
    frame_idx = 0
    while cap.isOpened():
        success, frame = cap.read()
        if not success:
            break
        h, w = frame.shape[:2]
        image_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = pose.process(image_rgb)

        if results.pose_landmarks:
            lm = results.pose_landmarks.landmark
            row = {
                "frame": frame_idx,
                "left_x": lm[15].x * w, "left_y": lm[15].y * h,
                "right_x": lm[16].x * w, "right_y": lm[16].y * h,
                "nose_x": lm[0].x * w, "nose_y": lm[0].y * h,
                "left_eye_x": lm[2].x * w, "left_eye_y": lm[2].y * h,
                "right_eye_x": lm[5].x * w, "right_eye_y": lm[5].y * h,
                "left_shoulder_y": lm[11].y * h, "right_shoulder_y": lm[12].y * h,
            }
            data.append(row)
        frame_idx += 1

    cap.release()
    return pd.DataFrame(data)

def load_openface_face_points(csv_path):
    df = pd.read_csv(csv_path)
    face_data = df[["frame", "x_8", "y_8", "x_27","y_27"]]  # 8=chin, 27=forehead
    face_data.columns = ["frame", "chin_x", "chin_y","forehead_x","forehead_y"]
    return face_data

def build_merged_df(pose_df, face_df):
    merged_df = pd.merge(pose_df, face_df, on="frame", how="inner")
    merged_df = merged_df.dropna()
    return merged_df

def detect_action_duration(merged_df, action_name, threshold=80, min_ratio=0.1):
    flags = []
    chin_threshold = 100
    away_threshold = chin_threshold + 30
    forehead_threshold = 90

    for _, row in merged_df.iterrows():
        # wrist distance
        left = lambda k: np.linalg.norm([row["left_x"] - row[f"{k}_x"], row["left_y"] - row[f"{k}_y"]])
        right = lambda k: np.linalg.norm([row["right_x"] - row[f"{k}_x"], row["right_y"] - row[f"{k}_y"]])
        # index distance
        left_index = lambda k: np.linalg.norm([row["left_index_x"] - row[f"{k}_x"], row["left_index_y"] - row[f"{k}_y"]])
        right_index = lambda k: np.linalg.norm([row["right_index_x"] - row[f"{k}_x"], row["right_index_y"] - row[f"{k}_y"]])

        d = {
            "left": {
                "chin": left("chin"),
                "nose": left("nose"),
                "forehead": left("forehead"),
                "eye": np.linalg.norm([row["left_x"] - row["left_eye_x"], row["left_y"] - row["left_eye_y"]])
            },
            "right": {
                "chin": right("chin"),
                "nose": right("nose"),
                "forehead": right("forehead"),
                "eye": np.linalg.norm([row["right_x"] - row["right_eye_x"], row["right_y"] - row["right_eye_y"]])
            },
        }

        if action_name == "chin_rest":
            chin_close = d["left"]["chin"] < chin_threshold or d["right"]["chin"] < chin_threshold
            eye_far = d["left"]["eye"] > away_threshold and d["right"]["eye"] > away_threshold
            nose_far = d["left"]["nose"] > away_threshold and d["right"]["nose"] > away_threshold
            flags.append(chin_close and eye_far and nose_far)

        elif action_name == "touching_face":
            touching = (
                any(d["left"][k] < threshold for k in ["chin", "eye", "nose"]) or
                any(d["right"][k] < threshold for k in ["chin", "eye", "nose"])
            )
            chin_rest_like = (
                (d["left"]["chin"] < chin_threshold or d["right"]["chin"] < chin_threshold) and
                (d["left"]["eye"] > away_threshold and d["right"]["eye"] > away_threshold) and
                (d["left"]["nose"] > away_threshold and d["right"]["nose"] > away_threshold)
            )
            flags.append(touching and not chin_rest_like)

        elif action_name == "hand_raise":
                #if the hand is higher than the shoulder
                left_up = row["left_y"] < row["left_shoulder_y"]
                right_up = row["right_y"] < row["right_shoulder_y"]

                hand_near_face = (
                    any(d["left"][k] < 100 for k in ["chin", "nose", "eye", "forehead"]) or
                    any(d["right"][k] < 100 for k in ["chin", "nose", "eye", "forehead"])
                )
                flags.append((left_up or right_up) and not hand_near_face)

        elif action_name == "support_forehead":
            flags.append(d["left"]["forehead"] < forehead_threshold or d["right"]["forehead"] < forehead_threshold)

        else:
            flags.append(False)

    chin_dists = [np.linalg.norm([row["left_x"] - row["chin_x"], row["left_y"] - row["chin_y"]]) for _, row in merged_df.iterrows()]

    count = sum(flags)
    ratio = count / len(flags) if flags else 0
    return {
        "action": action_name,
        "detected": ratio >= min_ratio,
        "frame_count": count,
        "ratio": round(ratio, 3)
    }

def analyze_hand_face_gestures(video_path, openface_csv):
    pose_df = extract_pose_wrist(video_path)
    face_df = load_openface_face_points(openface_csv)
    merged_df = build_merged_df(pose_df, face_df)

    results = [
        detect_action_duration(merged_df, "chin_rest", threshold=100, min_ratio=0.1),
        detect_action_duration(merged_df, "touching_face", threshold=110, min_ratio=0.1),
        detect_action_duration(merged_df, "support_forehead", threshold=90, min_ratio=0.1),
        detect_action_duration(merged_df, "hand_raise", min_ratio=0.03),
    ]
    return results
```

```python
def analyze_full_clip(video_path, keypoint_folder, segment_id=None):
    result = {}

    # basic information
    start, end = extract_segment_time_from_filename(os.path.basename(video_path))
    fps = get_fps(video_path)
    segment = (start, end)
    result['segment_id'] = segment_id or f"{os.path.basename(os.path.dirname(video_path))}_{os.path.basename(video_path)}"
    result['duration'] = end - start

    # head analysis(noding, shaking , movement)
    behavior = extract_behavior_features(video_path, keypoint_folder)
    if behavior:
        result.update(behavior)

    # AU analysis
    au_result = extract_and_analyze_au(video_path, output_dir="/content/openface_output", fps=fps)
    result.update(au_result)
    csv_file = os.path.basename(video_path).replace("mp4","csv")
    openface_csv = f"/content/openface_output/{csv_file}"


    # hand-face gesture
    gesture_results = analyze_hand_face_gestures(video_path, openface_csv)
    for gesture in gesture_results:
        action = gesture['action']
        result[f"{action}_detected"] = gesture['detected']
        result[f"{action}_frame_ratio"] = gesture['ratio']

    return result
```

```python
import cv2
from PIL import Image
import numpy as np
import torch
import cv2
from PIL import Image
import numpy as np
import torch

def extract_clip_embedding_from_segment(video_path, start, end, model, preprocess, device, num_frames=6, pooling="mean"):
    cap = cv2.VideoCapture(video_path)
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    total_frames = int((end - start) * fps)
    frame_indices = [int(start * fps + i * total_frames / num_frames) for i in range(num_frames)]

    images = []
    for idx in frame_indices:
        cap.set(cv2.CAP_PROP_POS_FRAMES, idx)
        success, frame = cap.read()
        if not success:
            continue
        image = Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        images.append(image)
    cap.release()

    if not images:
        return None

    processed_images = []
    for image in images:
        processed_image = preprocess(image).unsqueeze(0)  # Add batch dimension
        processed_images.append(processed_image)

    # Stack all preprocessed images and move to device
    images_tensor = torch.cat(processed_images, dim=0).to(device)

    with torch.no_grad():
        # Get embeddings as tensor (stay on GPU)
        embeddings = model.encode_image(images_tensor)

        # Perform pooling on GPU
        if pooling == "mean":
            pooled_embedding = torch.mean(embeddings, dim=0)
        elif pooling == "max":
            pooled_embedding = torch.max(embeddings, dim=0)[0]  # torch.max returns (values, indices)
        elif pooling == "none":
            pooled_embedding = embeddings
        else:
            raise ValueError("Invalid pooling method. Use 'mean', 'max', or 'none'")

    # Only convert to numpy at the very end
    return pooled_embedding.cpu().numpy()
```
